```python
# modelo inicial com 3 derivadas
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

k1= 0.15372
k2=0.8463
k3=1
kx=0.65
k4=0.35

tempo = np.arange (0,10,0.001)
lista1=[]

def eqidif(Y,T):
    E=Y[0]
    S=Y[1]
    F=Y[2]
    
    de_dt=-k1*E-k2*E
    ds_dt= k2*E-k3*S+kx*F
    df_dt= k3*S-k4*F-kx*F
    return [de_dt,ds_dt,df_dt]


# condição inicial
ce=3
cs=0.005
cf=0.0025



sol= odeint(eqidif,[3,2.25,2],tempo)

plt.plot(tempo,sol[:,0],'g',label="Estômago",linestyle='--')
plt.plot(tempo,sol[:,1],'b',label="Sangue", linewidth=3)
plt.plot(tempo,sol[:,2],'orange',label="Fígado")
plt.ylabel('Concentração de Anabolizantes (mg/L)')
plt.xlabel('Tempo (Dias)')
plt.legend(bbox_to_anchor=(0.7,0.95),loc=2,borderaxespad=0.)
plt.show() 


```
